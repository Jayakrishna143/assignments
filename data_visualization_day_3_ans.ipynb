##### import tips dataset.

```python
import pandas as pd,seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/tips.csv')
df.head()
```

##### 1. Write a Python program to display a bar chart of the total bill of Gender and write the intuition about the plot.

```python
### code here
sns.barplot(data = df,x = 'sex',y = 'total_bill')
# On average male paid more bill compared to Female
```

###### 2. Write a Python program to display a box plot of the total bill of day and hue = "smoker" and write the intuition about the plot.

```python
#### code here
sns.boxplot(data = df, x = 'day',y = 'total_bill',hue ='smoker')
# days such as sunday and saturday have higher overall bills and among gender male spends more compared to female on average
```

##### 3. Write a Python program to display a violin plot of the total bill of Day and hue = "sex" and write the intuition about the plot.

```python
### code here
sns.violinplot(data = df, x = 'day',y = 'total_bill',hue ='smoker')
# days such as sunday and saturday have higher overall bills and among gender male spends more compared to female on averagef
```

##### 4. Write a Python program to display a strip plot of the total bill of Day and hue = "sex" and write the intuition about the plot.

```python
### code here
sns.stripplot(data = df, x = 'day',y = 'total_bill',hue = 'sex')
# days such as sunday and saturday have higher overall bills and among gender male spends more compared to female on average
```

##### 5. Write a Python program to display heat map of dataframe and write the intuition about the plot.

```python
### code here
sns.heatmap(data = df[['total_bill','tip','size']].corr(),annot = True)
# Heatmap gives correlation among total_bill,tip and size all are positively correlated. while total_bill and size are more likely correlated compared to tip and size
```

##### 6. Write a Python program to display scatter plot of total bill of tip and write the intuition about the plot.

```python
### code here
sns.scatterplot(data = df, x = 'tip', y = 'total_bill',hue ='sex' )
# In the below scatterplot the relation between tip and total_bill is positively correlated and when hue is added male are more likely to give higher tips compared to female and same happens in total_bill
```

##### 7. Write a Python program to display density plot of all numerical columns from the dataframe and write the intution about plot.

```python
### code here
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for i,col in enumerate(['total_bill','tip','size']):
    sns.histplot(data = df ,x = col,ax = axes[i],kde  = True)
# all the graphs fallows normal distribution but right sqeaued
```

##### 8. Write a Python program to calculate average total bills by days and display top 4 days by plot.

```python
df.groupby('day')['total_bill'].mean().reset_index()
```

```python
### code here
df1 =  df.groupby('day')['total_bill'].mean().reset_index()
sns.barplot(data = df1, x = 'day',y = 'total_bill')
# sat and sunday has more avg total_bills
```

##### 2. What is box plot ?

A box plot, also known as a box-and-whisker plot, is a graphical representation of distribution of numerical data

1.  Minimum: The smallest value in the data set.
2.  First Quartile  The value that separates the lowest 25% of the data from the rest.
3.  Median  The middle value of the data set.
4.  Third Quartile  The value that separates the highest 25% of the data from the rest.
5.  Maximum: The largest value in the data set.

##### 3. What are the plots we can use for univeriate analysis.

For univariate analysis, plots like
*  histograms,
*  density plots (KDE),
*  box plots, and violin plots are used

##### 4. What are the plots we can use for bivariate analysis.




For bivariate analysis, plots like:

*   Scatter plots.
*   Line plots
*   Density plots
*   Box plots

*   Bar plots

###### 5. What are plot we can use for multivariate analysis.




For multivariate analysis, plots like:

*   Pair plots
*   Scatter plot matrix
*   3D scatter plots
*   Heatmaps (for correlation matrices)
*   Bubble plots
